```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv("Books.csv")
users = pd.read_csv("Users.csv")
ratings = pd.read_csv("Ratings.csv")
```

```python
books.head()
```

```python
ratings_with_name = ratings.merge(books,on = 'ISBN')
```

```python
ratings_with_name
```

##  Filtering data by removing those users who have rated less then 200 books

```python
filtered_1 = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
acceptable_users = filtered_1[filtered_1].index
```

```python
filtered_users = ratings_with_name[ratings_with_name['User-ID'].isin(acceptable_users)]
```

```python
filtered_2 = filtered_users.groupby('Book-Title').count()['Book-Rating'] >= 50
acceptable_books = filtered_2[filtered_2].index
```

```python
filtered_users_with_filtered_books = filtered_users[filtered_users['Book-Title'].isin(acceptable_books)] 
```

```python
filtered_users_with_filtered_books
```

```python
pt = filtered_users_with_filtered_books.pivot_table(index='Book-Title',columns='User-ID',values ='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

## using Cosine Similarity to find the similarity scores between all 706 books


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def Recommendor(book_name):
    index = np.where(pt.index==book_name)[0][0]
    suggestions = sorted(list(enumerate(similarity_scores[index])),key =lambda x:x[1],reverse=True)[1:6]

    for i in suggestions:
        print(pt.index[i[0]])
```

```python
Recommendor('You Belong To Me')
```

### Using k-Nearest Neighbour for clustering 

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
model.fit(pt)

distance, suggestion = model.kneighbors(pt.iloc[702,:].values.reshape(1,-1), n_neighbors=6 )




for i in suggestion[0][1:]:
    print(pt.index[i])
```
