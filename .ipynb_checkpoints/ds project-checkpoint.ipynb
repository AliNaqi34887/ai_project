```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
 ratings.head(30)
```

```python
ratings['User-ID'].unique().shape
```

#accepting user who have rated more than 200 books


```python
x = ratings['User-ID'].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['User-ID'].isin(y)]
```

```python
ratings.head()
```

#merging user and ratings via isbn

```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
ratings_with_books.head()
```

#how many individuals rated a specific book 

```python
number_rating = ratings_with_books.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={number_rating.columns[1]:'number_of_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
final_ratings = ratings_with_books.merge(number_rating, on='Book-Title')
```

```python
final_ratings.head(2)
```

#removing those books which are given less than 50 ratings

```python
final_ratings = final_ratings[final_ratings['number_of_rating']>=50]
```

```python
final_ratings.head(100)
```

```python
mx = final_ratings['number_of_rating'].max()
mx
```

```python
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['User-ID','Book-Title'], inplace=True)
```

```python
final_ratings.shape
```

#creating matrix using pivot table for clustering model


```python
matrix = final_ratings.pivot_table(columns='User-ID', index= 'Book-Title', values= 'Book-Rating')
```

```python
matrix
```

```python
matrix.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(matrix)
```

```python
def Recommendor(book_name):
    index = np.where(matrix.index==book_name)[0][0]
    suggestions = sorted(list(enumerate(similarity_scores[index])),key =lambda x:x[1],reverse=True)[1:6]

    for i in suggestions:
        print(matrix.index[i[0]])
```

```python
Recommendor('Harry Potter and the Chamber of Secrets (Book 2)')
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt



# Split the data into training and testing sets
matrixArray = np.array(matrix)
X_train, X_test, y_train, y_test = train_test_split(matrixArray, matrixArray, test_size=0.2, random_state=42)

# Train a KNN classifier
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)

# Generate predictions on the testing set
y_pred_proba = knn_classifier.predict_proba(X_test)[:, 1]  # Use predicted probabilities for the positive class

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calculate AUC (Area Under the Curve)
auc = roc_auc_score(y_test, y_pred_proba)

# Plot ROC curve
plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
